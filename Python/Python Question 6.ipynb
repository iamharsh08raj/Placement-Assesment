```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("pokemon_data.xlsx")
```

```python
df[0:10]
```

# ● Get all Pokemons whose spawn rate is less than 5%


```python
df1 = df[df['spawn_chance']<5]
```

```python
df1 = df1[['name','spawn_chance']]
```

```python
print(f"""all Pokemons whose spawn rate is less than 5% are\n\n{df1}""")
```

# ● Get all Pokemons that have less than 4 weaknesses

```python
df2 = df[df['weakness'].str.count(',')<3]
```

```python
df2 = df2[['name', 'weakness']]
```

```python
print(f"""all Pokemons that have less than 4 weaknesses are\n\n{df2}""")
```

# Get all Pokemons that have no multipliers at all

```python
df3 = df[df['multipliers'].isnull()]
```

```python
df3 = df3[['name', 'multipliers']]
```

```python
print(f"""all Pokemons that have no multipliers at all are\n\n{df3}""")
```

# Get all Pokemons that do not have more than 2 evolutions

```python
df4 = df[df['next_evolution'].str.count(',')<2]
```

```python
df4['evolution_count'] = df['next_evolution'].str.count(',')
```

```python
df4 = df4[['name', 'evolution_count', 'next_evolution' ]]
```

```python
print(f"""all Pokemons that do not have more than 2 evolutions are\n\n{df4}""")
```

# Get all Pokemon who have more than two types of capabilities

```python
df5 = df[df['type'].str.count(',')>2]
```

```python
df5['capabilities_count'] = df5['next_evolution'].str.count(',')
```

```python
df5 = df5[['name', 'capabilities_count', 'type']]
```
