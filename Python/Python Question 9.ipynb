```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD')
```

```python
df
```

# Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
non_clean_cars = df[df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] != "Clean Alternative Fuel Vehicle Eligible"][["Make", "Model"]]
```

```python
non_clean_cars
```

# Get all TESLA cars with the model year, and model type made in Bothell City.

```python
tesla_cars_bothell = df[(df["Make"] == "TESLA") & (df["City"] == "Bothell")][["Model Year", "Model"]]
```

```python
tesla_cars_bothell
```

# Get all the cars that have an electric range of more than 100, and were made after 2015

```python
electric_cars_with_range = df[(df["Electric Range"] > 100) & (df["Model Year"] > 2015)]
```

```python
electric_cars_with_range
```

# Draw plots to show the distribution between city and electric vehicle type

```python
plt.figure(figsize=(12, 8))
df.groupby(["City", "Electric Vehicle Type"]).plot(kind="bar", stacked=True)
plt.title("Distribution of Electric Vehicle Type by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```
